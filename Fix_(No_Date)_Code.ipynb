```python
import pandas as pd

# Load the data
df = pd.read_csv('/content/try_again_pt_100.csv')

# Replace "(no date)" with a blank string in the 'year' column
df['year'] = df['year'].replace('(no date)', '', regex=False)

# Display the modified DataFrame (optional, but good for verification)
display(df.head())
```

```python
# Save the modified DataFrame to a new CSV file
df.to_csv('cleaned_date_and_special_characters.csv', index=False)

print("DataFrame saved to 'cleaned_date_and_special_characters.csv'")
```
